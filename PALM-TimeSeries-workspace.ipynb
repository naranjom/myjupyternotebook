```python
time = dataset.variables['time']
#    Total kinetic energy of the flow, E = 0.5 * (u2+v2+w2), (3D domain average)
E = dataset.variables['E']
#    Resolved-scale turbulence kinetic energy of the flow (3D domain average)
Es = dataset.variables['E*']
##
dt = dataset.variables['dt']
#    Friction velocity (horizontal average)
us = dataset.variables['u*']
#    Vertical velocity scale of the CBL (horizontal average)
ws = dataset.variables['w*']
#    Temperature scale (constant flux layer), defined as w"pt"0 / u* (horizontal average)
ths = dataset.variables['th*']
#    Max u/v/w-component of the velocity
umax = dataset.variables['umax']
vmax = dataset.variables['vmax']
wmax = dataset.variables['wmax']
#    Divergence of the velocity field after the pressure solver has been called 
#    (normalized with respect to the total number of grid points)
div_new = dataset.variables['div_new']
#    Divergence of the velocity field before the pressure solver has been called 
#    (normalized with respect to the total number of grid points)
div_old = dataset.variables['div_old']
#    Height of the convective boundary layer (horizontal average) determined by the height
#    of the minimum sensible heat flux
z_i_wpt = dataset.variables['z_i_wpt']
#    Height of the convective boundary layer (horizontal average) determined by the temperature
#    profile, following the criterion of Sullivan et al. (1998)
z_i_pt = dataset.variables['z_i_pt']
#    Subgrid-scale sensible heat flux at k=0 (horizontal average) within the constant flux layer
w_pp_pt_pp0 = dataset.variables['w"pt"0']
#    Subgrid-scale heat flux (horizontal average) for z = zw(1)
w_pp_pt_pp = dataset.variables['w"pt"']
#    Total heat flux (horizontal average) for z = zw(1)
wpt = dataset.variables['wpt']
#    Subgrid-scale momentum flux (u-component) at k=0 (horizontal average) within the constant flux layer
w_pp_u_pp0 = dataset.variables['w"u"0']
#    Subgrid-scale momentum flux (v-component) at k=0 (horizontal average) within the constant flux layer
w_pp_v_pp0 = dataset.variables['w"v"0']
#    Subgrid-scale humidity flux at k=0 (horizontal average) within the constant flux layer, zero values
#    are output if humidity is not used
w_pp_q_pp0 = dataset.variables['w"q"0']
#    Potential temperature at the surface (horizontal average)
pt0 = dataset.variables['pt(0)']
#    Potential temperature for z = zu(1) (horizontal average)
ptz_mo = dataset.variables['pt(z_mo)']
#    Monin-Obukhov length
ol = dataset.variables['ol']
#    Subgrid-scale vertical scalar concentration flux (requires passive_scalar = .T.)
w_pp_s_pp = dataset.variables['w"s"']
```
