```python
#----axis = 1 is rows-------------------------------------------#
#----axis = 0 is columns-------------------------------------------#
data = np.loadtxt(fname='./swCarp/inflammation-01.csv', delimiter=',')
if data.max(axis=0)[0] == 0 and data.max(axis=0)[20] == 20:
    print("Suspicious looking maxima!")
elif data.min(axis=0).sum() == 0:
    print ("Minima add up to zero!")
else:
    print ("Seems OK!")
print("--------------------------\n")
print("Column 1:\n",data[:,0],"\n")
print("Column 21:\n",data[:,20],"\n")
print("Column 40:\n",data[:,39],"\n")

print("Column maxes:\n",data.max(axis=0),"\n")

print("Matrix shape:\n", np.shape(data))
```

```python
def fahr_to_kelvin(temp):
    return fahr_to_cels(temp) + 273.15

def fahr_to_cels(temp):
    return ((temp - 32) * (5/9))

def kelvin_to_celsius(temp):
    return temp - 273.15

print("The freezing point of water:", fahr_to_kelvin(32))
print("\nThe boiling point of water:", fahr_to_kelvin(212))
print("\nThe absolute zero in Celsius:", kelvin_to_celsius(0.0))
```

```python
numbers = [1.5, 2.3, 0.7, -0.001, 4.4]
total = 0.0
for n in numbers:
    assert n > 0.0, 'Data should only contain positive values'
    total += n
print("total is:", total)
```
